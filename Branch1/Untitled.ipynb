```python
l = ['Jan 2000 - Jun 2001', 'Jun 1997 - Nov 2030']
def date_conv(date):
    dates = []
    for x in date:
        d = datetime.datetime.strptime(x, '%b %Y - %b %Y')
        dates.append(d)
    return dates

dates = date_conv(l)
print(type(dates))    
print(dates)
```
